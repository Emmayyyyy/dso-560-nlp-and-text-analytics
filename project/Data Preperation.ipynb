# Occasion
## Data cleaning

```python
occ = tagged_full[tagged_full.attribute_name=='occasion']
occ.head(3)
```

```python
# combine same cateogries
occ['attribute_value'] = occ['attribute_value'].apply(lambda x: x.lower())
occ['attribute_value'] = occ['attribute_value'].apply(lambda x: 'daytonight' if x=='day to night' else x)
occ['attribute_value'] = occ['attribute_value'].apply(lambda x: 'nightout' if x=='night out' else x)
occ.head(3)
```

```python
# strip special characters for brand
occ['brand'] = occ['brand'].apply(lambda x: "".join([c.lower() for c in x if c.isalnum()]))
occ['brand'].nunique()
```

```python
# fill in missing value
occ['description'] = occ['description'].fillna("UNKNOWN_TOKEN")
occ['details'] = occ['details'].fillna("UNKNOWN_TOKEN")
```

```python
def detail_regex(line):
    t = line.replace("\n","")
    t1 = re.sub(r'\w*\d*(?:\s*)=(?:\s*)\d*\w*(?:-\w*)(?:\d*\w*)','SIZE_VALUE',t)
    t2 = re.sub(r'\d+\%','PERCENTAGE_TOKEN', t1)
    t3 = re.sub(r'(?:()\d*(cm)(?:/\d*)(?:))','HEIGHT_VALUE',t2)
    #t3 = re.sub(r'.*\(*(cm)(?:/\d*)\)*','HEIGHT_VALUE',t2)
    t4 = re.sub(r'\d*\'\d*"(?:\/\d*\w*)','HEIGHT_VALUE',t3)
    t5 = re.sub(r'\d*=\d*','SIZEVALUE',t4)
    t6 = re.sub(r'\d*\"\/\w*','MEASUREMENT_VALUE', t5)
    return t6
```

```python
occ['details']=occ['details'].apply(detail_regex)
```

## Collocation 
### name: freq=25

```python
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize, tokenwrap, bigrams
lemmatizer = WordNetLemmatizer()
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english') + [".",'.', ",",":", "''", "'s", "'", "``", "(", ")", "-"])
```

```python
name_list = []
for line in occ.name:
    line = line.replace("\n", "")
    if len(line) > 0:
        #line = [lemmatizer.lemmatize(token) for token in word_tokenize(line)]
        line = word_tokenize(line)
        name_list.append(line)
```

```python
collocation_finder = BigramCollocationFinder.from_documents(name_list)
collocation_finder.apply_freq_filter(25)
measures = BigramAssocMeasures()

col = collocation_finder.nbest(measures.raw_freq, 1000)
#collocation_finder.nbest(measures.likelihood_ratio, 80)
len(col)
```
